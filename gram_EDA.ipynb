```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data= pd.read_csv('gram_data.csv')
data.head()
```

```python
data.shape
```

### data preprocessing for EDA

```python
data.price=data.price.astype(float)      #changing price variable to float
```

```python
data['day']=data['date'].str.split(' ').str[0]
data['month']=data['date'].str.split(' ').str[1]        # splitting date string to three variables
data['year']=data['date'].str.split(' ').str[2]
data.head()
```

```python
data=data.drop('date',axis=1)                 #dropping the original date string
data.head()
```

```python
data[['day','year']]=data[['day','year']].astype(float)
```

```python
data.to_csv('gram_df.csv', index=False)
```

## Exploratory Data Analysis

### Fig.1

```python
fig, scatter = plt.subplots(figsize = (26,18))
sns.countplot(data.district, hue=data.variety)
```

### Fig.2

## INSIGHT:
### 1. In the first fig it is clear that Kolar district has maximum number of markets among all and Raichur and          Madikeri the least. Also count for hybrid variety is very less compared tomato variety. 
### 2. The same way market for hybrid variety is also very less when compared to tomato variety

### Fig.3

```python
fig, scatter = plt.subplots(figsize = (26,18))
sns.countplot(data.district, hue=data.market)
```

### Fig.4

```python
fig, scatter = plt.subplots(figsize = (22,18))
sns.barplot(data.district,data.price)
```

## INSIGHT:
### 1. From Fig.3 it is clear that Belgaum and Gadag has low number of markets and from Fig.4 we can see that the price of tomato is very huge in both these districts. It may be due factor such has low number of market causes demand-supply issues which inturn cause price increase.
### 2. The same way the prices of tomato in Kolar and Bangalore is balanced due to presence of huge number of markets in both these districts.
